```python
import numpy as np
mean_pitch_list=[]
mean_yaw_list=[]
for i in range(len(df)):
    if len(mean_pitch_list)<10:
        mean_pitch_list.append(df['pitch_eye'].loc[i])
    else:
        if df['pitch_eye'].loc[i]<100:
            _=mean_pitch_list.pop(0)
            mean_pitch_list.append(df['pitch_eye'].loc[i])

    if len(mean_yaw_list)<10:
        mean_yaw_list.append(df['yaw_eye'].loc[i])
    else:
        if df['yaw_eye'].loc[i]<100:
            _=mean_yaw_list.pop(0)
            mean_yaw_list.append(df['yaw_eye'].loc[i])

    
    if df['pitch_eye'].loc[i]>100:
        df.loc[i,'pitch_eye']=np.mean(mean_pitch_list)
    if df['yaw_eye'].loc[i]>100:
        df.loc[i,'yaw_eye']=np.mean(mean_yaw_list)

```
